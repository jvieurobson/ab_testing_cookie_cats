The following project uses a dataset provided by Tactile Games - a casual social game developer - as the result of an A/B testing campaign they ran within their game Cookie Cats, to evaluate the effect of modifying a game feature.

![alternatvie text](https://tactilegames.com/wp-content/uploads/2018/05/cookie-cats.png)

Cookie Cats is a hugely popular mobile puzzle game developed by Tactile Entertainment. It's a classic "connect three"-style puzzle game where the player must connect tiles of the same color to clear the board and win the level.  As players progress through the levels of the game, they will occasionally encounter gates that force them to wait a non-trivial amount of time or make an in-app purchase to progress. In addition to driving in-app purchases, these gates serve the important purpose of giving players an enforced break from playing the game, hopefully resulting in that the player's enjoyment of the game being increased and prolonged. 

Based on previous data they have collected - user interviews, user behavior - Tactile Games will have generated two hypotheses.
1. Moving gate locations to a later level will increase the total number of game_rounds played by each user
2. Moving gate locations to a later level will maximize the likelihood that users return to play 1 and 7 days after installation. 


Both of these hypotheses serve to investigate whether moving the gate will maximize the likelihood of in-game purchases.

We don't have enough information to determine which variable - total number of games played by each user, or number of players returning 1 and 7 days after installation - will have the greatest impact on the monetary value of in-game purchases. It is possible that a certain portion of users play many games within the first few days after installation, but quickly stop after those first few days. Alternatively, players could play only a few game rounds distributed across 1 to 7 days of activity.

They have then ran an A/B test in which two samples have been generated from their user base and divided into a control group - users exposed to a gate placement at level 30, and the treatment group - users exposed to a gate placement at level 40. 

They have then recorded the total number of games played by each user during the 14-day period after installation, and whether users come back 1 and 7 days after installation.

Both the sum_gamerounds for the control (gate_30) and treatment group are log-normally distributed.
We therefore have a few options to pursue in order to complete a statistical test. 

We know that:
a) Two samples. The data collected is the product of two random samples.
b) Independently sampled. The two samples' observations are independent of each other - i.e. no autocorrelation. No users are found present twice or more in both the gate_30 (control) and gate_40 (treatment) groups.
b) One categorical and one numeric variable. We are investigating the relationship between gate placement (categorical) and total number of gamerounds played per user (numeric).
d) Parametric. Our data is normally distributed, and since we would like to maximize our Power, we will be using a parametric statistical test. 

Unknowns:
a) Sample variances. If the variance of both sample groups are approximately equivalent, we can perform a Student's t-test for 2 independent samples.

```python
'group_30 and group_30 sum_gamerounds variance'
print("gate_30 users sum_gamerounds variance: ", round(statistics.variance(gate_30.sum_gamerounds), 1))
print("gate_40 users sum_gamerounds variance: ", round(statistics.variance(gate_40.sum_gamerounds), 1))
```

Since the variances of the two sample groups are approximately equivalent, we can perform a Student's t-test for 2 independent samples.

Alternatively, we could perform a bootstrap statistical test.
